```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# for city in cities:
#     print(city)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = ["metric", "imperial"]

# empty lists to populate data into fields

cloudieness = []
country = []
date = []
humidity = []
max_temp = []
lat = []
lng = []
wind_speed = []


city = "praya"
    
query_url = f"{url}appid={api_key}&q={city}&units={units[0]}"

# print(query_url)
    
city_response = requests.get(query_url)
city_json = city_response.json()
    
pprint(city_json)
    
```

```python
# subcities = cities[0:50]
cityCounter = 0
city_in_df = []

for city in cities:
    time.sleep(1)
    cityCounter+=1
    
    query_url = f"{url}appid={api_key}&q={city}&units={units[0]}"
    
    city_response = requests.get(query_url).json()

#     city_json = city_response.json()

#     pprint(city_json)
#     print(city)


    try:
        
        cloudieness.append(city_response['clouds']['all'])
        country.append(city_response['sys']['country'])
        date.append(city_response['dt'])
        humidity.append(city_response['main']['humidity'])
        max_temp.append(city_response['main']['temp_max'])
        lat.append(city_response['coord']['lat'])
        lng.append(city_response['coord']['lon'])
        wind_speed.append(city_response['wind']['speed'])
        
        print(f"Processing Record {cityCounter} of Set 1 | {city}")
        
        city_in_df.append(city)
    except:
        
        print(f"Record {cityCounter}, city of {city} not found. Skipping...")
    
    
print(len(cloudieness))
print(len(country))
print(len(lat))
print(len(lng))
print(len(date))
print(len(humidity))
print(len(max_temp))
print(len(wind_speed))
print(len(city_in_df))

#     lat.append(response['coord']['lat'])
#     temp.append(response['main']['temp'])
```

```python
weather_df = pd.DataFrame({
    "City": city_in_df,
    "Country": country,
    "Lat": lat,
    "Lng": lng,
    "Date": date,
    "Cloudieness": cloudieness,
    "Humidity": humidity,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed
})

weather_df.to_csv("output_data/weather_df.csv", index=False, header=True)
weather_df.head()
```
